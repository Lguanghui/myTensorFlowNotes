# sort/argsort

```
a=tf.random.shuffle(tf.range(5))
a
```

```
tf.sort(a,direction='DESCENDING')#降序
```

```
tf.sort(a,direction="ASCENDING")#升序
```

```
tf.argsort(a,direction="ASCENDING")#得到升序排列后的索引
```

```
tf.argsort(a,direction="DESCENDING")#得到降序排列后的索引
```

# Top_k（最大/最小前k个）

```
b=tf.constant([[3,2,1,5],[2,1,4,6]])
```

```
bb=tf.math.top_k(b,2)
#返回结果是两个，Values是值，indices是索引
```

```
bb.indices,bb.values
```
