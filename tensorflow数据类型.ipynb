# tensor的属性

```
#当前tensor运行的设备（cpu/gpu）
with tf.device('cpu'):
  a=tf.constant([1])
with tf.device('gpu'):
  b=tf.range(4)
print(a.device)
print(b.device)
```

```
#cpu和gpu转换
aa=a.gpu()
aa.device
```

```
#转化为numpy类型
bb=b.numpy()
bb
```

```
#查看维度
bb.ndim
```

```
#判断是否为tensor
tf.is_tensor(b) , tf.is_tensor(b.numpy())
```

```
#判断数据类型
b.dtype==tf.double
```

#数据类型转化

```
a=np.arange(5)
a.dtype
```

```
#转化为tensor
aa=tf.convert_to_tensor(a)
aaa=tf.convert_to_tensor(a,dtype=tf.int32)
aa,aaa
```

```
#更改tensor的数据类型
tf.cast(aa,dtype=tf.double)
```

```
#整型和bool的转换
i=tf.constant([0,1])
tf.cast(i,dtype=tf.bool)
```

#Variable类型

```
在TensorFlow中，变量(Variable)是特殊的张量(Tensor)，它的值可以是一个任何类型和形状的张量。
与其他张量不同，变量存在于单个 session.run 调用的上下文之外，也就是说，变量存储的是持久张量，当训练模型时，用变量来存储和更新参数。除此之外，在调用op之前，所有变量都应被显式地初始化过。
```

```
v = tf.Variable([1,2,3])   #创建变量v，为一个array
print(v)  #查看v的shape，不是v的值。结果是： <tf.Variable 'Variable:0' shape=(3,) dtype=int32_ref>
z = tf.Variable(tf.zeros([20]))  #初始化一个形状为[20]的张量， 里面的元素值全部为0.
```
