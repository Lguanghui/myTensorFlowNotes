```
import tensorflow as tf
```

# 加减乘除，取余，整除

```
a=tf.fill([2,2],2.)
b=tf.ones([2,2])
```

```
a+b,a-b
```

```
a*b,a/b   #均是对应元素进行操作
```

```
b//a #整除，除后取整数
```

```
a%b #取余
```

# log对数和exp指数运算

如果要实现非e为底数的log运算，则需要用log(x)/log(y)

```
tf.math.log(a)
```

```
tf.math.exp(b)
```

# 幂次运算和开放运算

```
tf.math.pow(a,4)
```

或者

```
a**4
```

```
tf.math.sqrt(a)
```

# 矩阵运算

```
a@b
```

或者

```
tf.matmul(a,b)
```

当多维张量使用矩阵运算时，会把张量的最后两维当作矩阵来运算

```
c=tf.random.normal([3,9,3,2])
d=tf.random.normal([3,9,2,3])
```

```
(c@d).shape
```

```
dd=tf.random.normal([2,3])
# 这时候直接c@dd是不可以的，tf不会自动使用隐式broadcast
ddd=tf.broadcast_to(dd,[3,9,2,3])
(ddd@c).shape
```
