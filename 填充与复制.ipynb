# 填充（Pad)

和安卓开发中的padding含义类似
image.png

tf.pad()中，有两层[]，其中，每一个维度都对应着一个[x,y],x代表在这个维度的前面padding，y代表在这个维度的后面padding

```
a=tf.reshape(tf.range(9),[3,3])
```

```
tf.pad(a,[[0,0],[0,0]])#都不padding
```

```
tf.pad(a,[[1,0],[0,0]])#第一行padding
```

```
tf.pad(a,[[1,1],[1,1]])#上下左右都padding
```

```
b=tf.random.normal([4,28,28,3])
```

```
bb=tf.pad(b,[[0,0],[2,2],[2,2],[0,0]])
bb.shape
```
