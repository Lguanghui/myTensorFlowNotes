```
import tensorflow as tf
import pandas as pd
import numpy as np
```

# 张量的合并（拼接）

## 不增加额外维度的方式（tf.concat）

假设[4,35,8],[2,35,8]分别表示班级1-4，班级5-6的学生成绩。想要把这两份成绩合并在一起

```
a=tf.ones([4,35,8])
b=tf.ones([2,35,8])
```

```
#保证要合并的维度外，其他各维度的大小必须一致
c=tf.concat([a,b],axis=0)
c.shape
```

## 额外增加维度的方式（tf.stack）

```
d=tf.ones([4,35,8])
e=tf.ones([4,35,8])
```

```
#要保证对应的维度大小一致
f=tf.stack([d,e],axis=0)
f.shape
```

```
ff=tf.stack([d,e],axis=2)
ff.shape
```

# 张量的分割

## 某一维度完全分割（tf.unstack）

unstack可以在任意轴上分割，能够返回张量列表，原来的维度会消失

```
dd,ee=tf.unstack(f,axis=0)
dd.shape,ee.shape
```

```
s=tf.ones([3,4,6])
ss=tf.unstack(s,axis=2)
ss[0].shape
```

## 按照指定形状/数量分割（tf.split）

```
v=tf.ones([3,4,5,6])
```

```
vv=tf.split(v,axis=3,num_or_size_splits=[2,3,1])
len(vv)
```

```
vv[0].shape,vv[1].shape,vv[2].shape
```
