##gather按索引收集

```
#假设c有4个班级，每个班级35个学生，每个学生有8门成绩
c=tf.ones([4,35,8])
```

收集第2、3号（从1号开始）班级的成绩

```
cc=tf.gather(c,axis=0,indices=[1,2])
cc.shape
```

上式等同于

```
c[1:3]
```

但是，gather的索引顺序是按indices里面的索引顺序来的。也就是说，indices[2,3,4]跟indices[3,4,2]得到的结果是不一样的

axis代表抽取的哪一个维度，例如下面代码表示在各个班级里按顺序抽取第4、6、1、23、4号同学的所有成绩

```
ccc=tf.gather(c,axis=1,indices=[4,6,1,23,4])
ccc.shape
```

##gather_nd在多个维度同时收集

如果我们想抽取1号班级2号学生，3号班级8好学生。。。的成绩，则需要在多个维度进行抽取

```
#抽取0号班级全部学生的成绩
cccc=tf.gather_nd(c,[0])
cccc.shape
```

```
#抽取0号班级1号学生的成绩
ccccc=tf.gather_nd(c,[0,1])
ccccc.shape
```

两级[ ]中，最外面的[ ]代表抽取结果的list，里面有多少个[ ]，就会有多少样本。第二级[ ]表示每个样本的索引

```
#抽取0号班级7号学生、1号班级4号学生的所有成绩
ccccccc=tf.gather_nd(c,[[0,7],[1,4]])
ccccccc.shape
```

```
#抽取0号班级1号学生、1号班级8号学生的某一门成绩
cccccc=tf.gather_nd(c,[[0,1,7],[1,8,7]])
cccccc.shape
```

下面代码跟上面代码差不多，但它多了一层[ ]，就是把结果放入了一个list中的list

```
ccccccc=tf.gather_nd(c,[[[0,1,7],[1,8,7]]])
ccccccc.shape
```
