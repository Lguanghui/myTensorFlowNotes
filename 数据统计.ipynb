```
import tensorflow as tf
```

# 向量的范数

向量的二范数

```
a=tf.ones([2,2])
```

```
tf.norm(a)
```

指定维度

```
tf.norm(a,axis=1)
```

向量的一范数

```
tf.norm(a,ord=1,axis=1)
```

# 张量某一维度最大值最小值平均值

```
b=tf.random.normal([4,10])
```

如果不加axis参数，那么求得的就是所有元素的最大值、最小值、均值

```
tf.reduce_max(b),tf.reduce_min(b),tf.reduce_mean(b)
```

指定维度后，计算最大值/最小值/均值，就会得到一个降维的结果

```
tf.reduce_max(b,axis=0)
```

```
tf.reduce_max(b,axis=1)
```

# 返回极值的索引

tf.argmax(vector, 1)：返回的是vector中的最大值的索引号，如果vector是一个向量，那就返回一个值，如果是一个矩阵，那就返回一个向量，这个向量的每一个维度都是相对应矩阵行的最大值元素的索引号。

```
tf.argmax(b,axis=1)
```

```
tf.argmin(b,axis=0)
```

# 张量比较

```
c=tf.ones([2,3])
d=tf.random.normal([2,3])
```

```
e=tf.equal(c,d)
```

可以求得匹配正确的个数

```
tf.reduce_sum(tf.cast(e,dtype=tf.int32))
```
