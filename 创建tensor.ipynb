```
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#tf.convert_to_tensor

```
a=tf.convert_to_tensor(np.ones([2,3],dtype=np.int32))
a
```

```
b=tf.convert_to_tensor([2,3])#整型
bb=tf.convert_to_tensor([2,3.])#浮点型,如果元素的数据类型不一致，则自动使用精度最高的类型
b,bb
```

#tf自带函数

```
c=tf.zeros([2,3],dtype=tf.float64)#[]里面是shape
c
```

##利用已有数据的shape创建tensor

```
cc=tf.zeros(c.shape)#使用缺省数据类型
ccc=tf.zeros_like(c)#使用相同的数据类型
cc,ccc
```

##创建任意相同值的矩阵

```
d=tf.fill([2,5],9)
d
```

#创建随机数矩阵

##创建正态分布的tensor

```
n=tf.random.normal([3,9],mean=1,stddev=1)
n
```

```
t=tf.random.normal([1000],mean=1,stddev=2)
plt.hist(t,density=True)#绘制直方图
```

```
#截断的正态分布
nn=tf.random.truncated_normal([2,9],mean=0,stddev=1)
nn
```

##均匀分布的tensor

```
m=tf.random.uniform([1000],minval=0,maxval=6)
plt.hist(m,density=True)#绘制直方图
```

##随机打散 Random Permutation

```
s=tf.range(10)
ss=tf.random.shuffle(s)
ss
```
