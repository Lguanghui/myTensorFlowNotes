# 概念

```
import tensorflow as tf
import pandas as pd
import numpy as np
```

参考链接：[tensorflow 2.0 基础操作 之 Broadcasting机制](https://blog.csdn.net/z_feng12489/article/details/89332012)

broadcast是张量维度扩张的手段，但是没有在数据层面上的复制。
是一种数据优化的手段。高效且直观。

当两个维度不同的张量相加时（例如，[4,32,32,3]+[3]）：


*   从低维开始匹配
*   如果当前维度dim=1，则扩张成相同的维度
*   如果没有当前维度，则插入一个维度，再扩张。
*   否则，不能使用broadcast



[4,32,32,3]+[3]：

[3]➡[1,1,13]➡[4,32,32,3]

# 练习

```
x=tf.random.normal([4,32,32,3])
y=tf.random.normal([3])
```

**在满足boradcast机制的条件时，tf会自动的隐式调用broadcast**

```
(x+y).shape
```

```
# z不满足机制条件
z=tf.random.normal([2,3])
z+x
```

显式调用broadcast

```
yy=tf.broadcast_to(y,x.shape)
yy.shape
```
